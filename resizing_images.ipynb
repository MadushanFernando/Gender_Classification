```python
from PIL import Image
import PIL
import os
import random
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
print("Train Data : ")
for dirpath, dirnames, filenames in os.walk("dataset/training_set"):
    print(f"There is {len(dirnames)} directories and {len(filenames)} in the path : {dirpath}")
print("Test Data : ")
for dirpath, dirnames, filenames in os.walk("dataset/test_set"):
    print(f"There is {len(dirnames)} directories and {len(filenames)} in the path : {dirpath}")
  

    
```

```python
trainingcats = []

for dirpath, dirnames, filenames in os.walk("dataset/training_set/cats"):
        for i in filenames:
            trainingcats.append(dirpath+'/'+i)
#print(trainingcats)


def view_random_images(image_set):
    plt.figure(figsize=(20,10))
    for i in range(5):
        image = random.choice(image_set)
        img = mpimg.imread(image)
        plt.subplot(5,1,i+1)
        plt.imshow(img)
view_random_images(image_set = trainingcats)
```

```python
trainingdogs = []
for dirpath, dirnames, filenames in os.walk("dataset/training_set/dogs"):
        for i in filenames:
            trainingdogs.append(dirpath+'/'+i)
view_random_images(image_set = trainingdogs)
```

```python
testingdogs = []
for dirpath, dirnames, filenames in os.walk("dataset/test_set/dogs"):
        for i in filenames:
            testingdogs.append(dirpath+'/'+i)
view_random_images(image_set = testingdogs)
```

```python
testingcats = []
for dirpath, dirnames, filenames in os.walk("dataset/test_set/cats"):
        for i in filenames:
            testingcats.append(dirpath+'/'+i)
view_random_images(image_set = testingcats)
```

# Resizing Images

```python
def resize_and_save_images(category,class_name,original_set):
    z=1
    for i in original_set:
        before_image=Image.open(i)
        after_image = before_image.resize((300,300))
        print("resized_images/"+category+"/"+class_name+"/"+str(z)+".jpg")
        after_image.save("resized_images/"+category+"/"+class_name+"/"+str(z)+".jpg")
        z=z+1
    
```

### Resizing all the test and training images

```python
resize_and_save_images(category='test',class_name='cats',original_set=testingcats)
```

```python
resize_and_save_images(category='test',class_name='dogs',original_set=testingdogs)
```

```python
resize_and_save_images(category='train',class_name='cats',original_set=trainingcats)
```

```python
resize_and_save_images(category='train',class_name='dogs',original_set=trainingdogs)
```
